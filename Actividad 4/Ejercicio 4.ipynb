-Plot histogram of DIC and TA

```python
def import_csv(file):
    # Read the file into a DataFrame: df
    import pandas as pd
    return pd.read_csv(file)
```

```python
# Import the file
df = import_csv("Terminos_lagoon_TA_DIC_2023_RawData.csv")
```

```python
print(df.columns)
```

```python
TA= df['ta_micromol_kg']
DIC = df['dic_micromol_kg']

```

```python
import matplotlib.pyplot as plt

plt.hist(TA, bins=20, alpha=0.5, label='TA')
plt.xlabel('TA (micromol/kg)')
plt.ylabel('Frequency')
plt.title('Histogram of Total Alkalinity (TA)')
plt.show()
```

```python
plt.hist(DIC, bins=20, alpha=0.5, label='DIC')
plt.xlabel('DIC (micromol/kg)')
plt.ylabel('Frequency')
plt.title('Histogram of Dissolved Inorganic Carbon (DIC)')
plt.show()
```
